```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your CSV file
file_path = r"C:\Users\tongu\OneDrive\Documents\1 Data Analyst Training Arc\11 SQL 2\SkillsTop10.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Count skill frequency
skill_counts = df['skills'].value_counts().reset_index()
skill_counts.columns = ['Skill', 'Count']

# Select top 10 skills
top_skills = skill_counts.head(10)

# Create a gradient color map
colors = plt.cm.Blues(np.linspace(1, .1, len(top_skills)))  # Blue gradient

# Plot
plt.figure(figsize=(10,6))
plt.barh(top_skills['Skill'], top_skills['Count'], color=colors)
plt.xlabel('Frequency')
plt.ylabel('Skill')
plt.title('Top 10 Skills in Top 10 Paying Data Analyst Jobs in California (2025)')
plt.gca().invert_yaxis()  # Highest count on top
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
file_path = r"C:\Users\tongu\OneDrive\Documents\1 Data Analyst Training Arc\11 SQL 2\SQL_Project_Data_Job_Analysis\Skills in Demand.csv"  # Update path if needed
df = pd.read_csv(file_path)

# Sort values for cleaner plotting
df_sorted = df.sort_values(by="demand_count", ascending=False)

# Pick a color map with many distinct colors
colors = plt.cm.tab20.colors  # tab20 provides 20 distinct colors

# If you have more than 20 skills, cycle through colors
bar_colors = [colors[i % len(colors)] for i in range(len(df_sorted))]

# Create the bar chart
plt.figure(figsize=(10,6))
bars = plt.bar(df_sorted["skills"], df_sorted["demand_count"], 
               color=bar_colors, width=0.4)  # width < 1 makes bars narrower

# Add labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 2, 
             str(height), ha='center', va='bottom', fontsize=9)

# Titles and labels
plt.title("Top Skills in Demand for Data Analysts in California (2025)")
plt.ylabel("Demand Count")
plt.xlabel("Skills")
plt.xticks(rotation=45, ha="right")
plt.xlim(-0.5, len(df_sorted) -.5 )

plt.tight_layout()
plt.show()
```
